```python
corpus = ["코로나 거리두기와 코로나 상생지원금 문의 입니다.",
          "지하철 운행시간과 지하철 요금 문의 입니다.",
          "지하철 승강장 문의 입니다.",
          "택시 승강장 문의 입니다."]

corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvect = CountVectorizer()
cvect.fit(corpus)
dtm = cvect.transform(corpus)

dtm
```

```python
dtm = cvect.fit_transform(corpus)

dtm

cvect
```

```python
cvect.vocabulary_
```

```python
vocab = cvect.get_feature_names_out()
vocab
```

```python
stop_words = ["코로나", "문의입니다"]

cvect= CountVectorizer(analyzer ='char', ngram_range=(1,5), min_df=2, max_df=1.0, max_features=30,stop_words=stop_words)
dtm = cvect.fit_transform(corpus)
vocab = cvect.get_feature_names_out()
df_dtm = pd.DataFrame(dtm.toarray(), columns=vocab)

df_dtm
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = ["코로나 거리두기와 코로나 상생지원금 문의 입니다.",
          "지하철 운행시간과 지하철 요금 문의 입니다.",
          "지하철 승강장 문의 입니다.",
          "택시 승강장 문의 입니다."]

tfidfvect = TfidfVectorizer()
tfidfvect.fit(corpus)
dtm = tfidfvect.transform(corpus)

dtm
```

```python
dtm = tfidfvect.fit_transform(corpus)
dtm

dtm.toarray()
```

```python
vocab = tfidfvect.get_feature_names_out()
df_dtm = pd.DataFrame(dtm.toarray(), columns=vocab)

print("단어수 : ", len(vocab))
print(vocab)

display(df_dtm.style.background_gradient())
```
